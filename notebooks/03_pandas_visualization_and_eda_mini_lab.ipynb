```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

sns.set(style="whitegrid")

df = pd.read_csv("../data/Lending-company.csv")
df["TotalPrice"] = df["TotalPrice"].fillna(df["TotalPrice"].median())
```

```python
plt.figure(figsize=(8,5))
sns.histplot(df["TotalPrice"], kde=True)
plt.title("Distribution of TotalPrice")
plt.xlabel("TotalPrice")
plt.ylabel("Frequency")
plt.show()
```

### Distribution Interpretation

- TotalPrice dağılımı simetrik değildir.
- Sağ kuyruk belirgindir (right-skewed).
- Fiyatların büyük çoğunluğu 16k–23k bandında yoğunlaşmıştır.
- 50k+ değerler nadir fakat dağılımı sağa doğru uzatmaktadır.

```python
plt.figure(figsize=(10,6))
sns.boxplot(x="Region", y="TotalPrice", data=df)
plt.xticks(rotation=45)
plt.title("TotalPrice by Region")
plt.show()
```

### Regional Comparison Interpretation

- Region’lar arasında medyan fiyat seviyeleri genel olarak benzerdir.
- Region 13 daha düşük medyan fiyat göstermektedir.
- Bazı region’larda yüksek değerli aykırı gözlemler (50k+) bulunmaktadır.
- Genel fiyat seviyesi bölgeler arasında dramatik farklılık göstermemektedir.

```python
plt.figure(figsize=(8,5))
sns.boxplot(x="LoanStatus", y="TotalPrice", data=df)
plt.title("TotalPrice by LoanStatus")
plt.show()
```

### Loan Status Interpretation

- LoanStatus grupları arasında medyan fiyat farkı dramatik değildir.
- Blocked krediler nispeten daha düşük fiyat bandında yoğunlaşmaktadır.
- Yüksek fiyatlı krediler (outliers) daha çok Active ve Finished Payment gruplarında görülmektedir.
- Bu durum kredi büyüklüğü ile ödeme davranışı arasında potansiyel bir ilişki olabileceğini düşündürebilir.
